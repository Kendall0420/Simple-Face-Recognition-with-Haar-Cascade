```
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```
def upload_files():
  from google.colab import files
  uploaded = files.upload()
  for k, v in uploaded.items():
    open(k, 'wb').write(v)
  return list(uploaded.keys())
```

```
def blur_img(upload_img):
    for i in upload_img:
      image = i
  
    # Read the image
    img = cv2.imread(image)
    # Greyscale the Image
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Load cascade
    cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

    # Detect faces, uses greyscaled image for easier edge detections
    faces = cascade.detectMultiScale(gray, 1.3, 5) # 1.3 as in scale factor from 100% down to 70%

    # Create blur effects for the faces
    for (x,y,w,h) in faces:
        img[y:y+h, x:x+w] = cv2.blur(img[y:y+h, x:x+w], (50,50))

    # Display image with blurred face
    cv2_imshow(img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```
# Load Image
uploaded_img = upload_files()
blur_img(uploaded_img)
```
